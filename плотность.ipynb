```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
from math import e, log
```

```python
def my_norm(x):
    return [(2 * e **(abs(i)) / (e ** (2 * abs(i)) + 1))for i in x]
def new_norm(x):
    return [(log((e**(i) + 1)/(e**(i) - 1))) for i in x]
```

```python
mu = 8
variance = 2 * mu
sigma = math.sqrt(variance)
x = np.linspace(0.00000001, 20, 100)
y = my_norm(x)
#plt.plot(x, y)
#plt.show()
y_ = stats.norm.pdf(new_norm(x), mu, sigma)
for i in range(len(x)):
     y[i] *= y_[i]
plt.plot(x, y, label = 'mu = 3', color = 'purple')


mu = 5
variance = 2 * mu
sigma = math.sqrt(variance)
x = np.linspace(0.00000001, 10, 1000)
y = my_norm(x)
#plt.plot(x, y)
#plt.show()
y_ = stats.norm.pdf(new_norm(x), mu, sigma)
for i in range(len(x)):
     y[i] *= y_[i]
plt.plot(x, y, label = 'mu = 3', color = 'gold')

mu = 3
variance = 2 * mu
sigma = math.sqrt(variance)
x = np.linspace(0.00000001, 10, 100)
y = my_norm(x)
#plt.plot(x, y)
#plt.show()
y_ = stats.norm.pdf(new_norm(x), mu, sigma)
for i in range(len(x)):
     y[i] *= y_[i]
plt.plot(x, y, label = 'mu = 5', color = 'red')

mu = 0.1
variance = 2 * mu
sigma = math.sqrt(variance)
x = np.linspace(0.00000001, 10, 100)
y = my_norm(x)
#plt.plot(x, y)
#plt.show()
y_ = stats.norm.pdf(new_norm(x), mu, sigma)
for i in range(len(x)):
     y[i] *= y_[i]
plt.plot(x, y, label = 'mu = 0.1', color = 'blue')

mu = 0.00001
variance = 2 * mu
sigma = math.sqrt(variance)
x = np.linspace(0.00000001, 10, 100)
y = my_norm(x)
#plt.plot(x, y)
#plt.show()
y_ = stats.norm.pdf(new_norm(x), mu, sigma)
for i in range(len(x)):
     y[i] *= y_[i]
plt.plot(x, y, label = 'mu = 0.00000001', color = 'green')

mu = 1000
variance = 2 * mu
sigma = math.sqrt(variance)
x = np.linspace(0.00000001, 10, 100)
y = my_norm(x)
#plt.plot(x, y)
#plt.show()
y_ = stats.norm.pdf(new_norm(x), mu, sigma)
for i in range(len(x)):
     y[i] *= y_[i]
plt.plot(x, y, label = 'mu = 0.0', color = 'black')
plt.show()
```
